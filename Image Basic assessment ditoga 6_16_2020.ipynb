# Question 2: correct the image in the right RGB order to display the natural color of the pic

```python
from PIL import Image

new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  #transform from RGB go BGR 

im2_triangle = new_img.copy() 
plt.imshow(new_img)
```

```python
# reference to flipping and rotating image
# https://note.nkmk.me/en/python-opencv-numpy-rotate-flip/

flipped_img = cv2.flip(new_img, 0) # flipped along horizontal axis
plt.imshow(flipped_img)
```

```python
# Question 6:  Create a script that opens the 
# picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.
```

```python
# This shows on a separate window using cv2.imshow(<window>, img)
# variable 
# To draw a circle, you need its center coordinates and radius. We will draw a circle inside the rectangle drawn above.
# cv.circle(img,(447,63), 63, (0,0,255), -1)
# reference: https://docs.opencv.org/master/dc/da5/tutorial_py_drawing_functions.html

img_w = cv2.imread('./original/Computer-Vision-with-Python/DATA/dog_backpack.jpg')
#print(type(img))
def draw_circle(event, x,y,flags,params):
    
    if event == cv2.EVENT_RBUTTONDOWN:
         cv2.circle(img_w,(440,700), 100, (0,0,255),10)
         # https://www.geeksforgeeks.org/python-opencv-cv2-circle-method/
         #  Syntax: cv2.circle(image, center_coordinates, radius, color, thickness)

#############################################################################################################
# Event Driven display and termination of the image 
# Show the image with OpenCV

cv2.namedWindow(winname = 'window_name')
# call back to the drawing_function
cv2.setMouseCallback('window_name',draw_circle)

while True:
    cv2.imshow('window_name',img_w)   # if I did that in plt.imshow('window_name,img'), then we will get a blueish color
    # Wait for something on keyboard to be pressed to close window.
    if cv2.waitKey(1)& 0xFF ==27:   # remember waitKey(<millisecond value>) and 27 is 'esc' button
        break
cv2.destroyAllWindows()
```
