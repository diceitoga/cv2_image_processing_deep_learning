# Edge Detection using Canny Edge Detector 7_17_2020

```python
'''
Edge Detection using:  Canny Edge Detector Developed in 1986
Steps to Edge detection
Step1:  Apply Gaussian filter to smooth it out and eliminate noise (noisy will detect too much)
    - find the intensity gradient of the image (you have to play with it)
Step2:  Apply non-maxumum supression to rid of suprious response
Step3:  Apply Double threshold to determine edges
Step4:  Canny Edge Detection Process: 
    - Track edges by 'Histeresis'
    - supress weak edges
    
we will do this step backwards in this example to show how each step without the other is enough from qualitative standpoint

***Caution:  for HD images, apply custom filter before using Canny Edge Detection


'''
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./original/Computer-Vision-with-Python/DATA/sammy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
threshold1 = 127
threshold2 = 127


edge = cv2.Canny(image = img, threshold1=threshold1, threshold2=threshold2)
plt.imshow(edge)
```

```python
threshold1 = 0
threshold2 = 255


edge = cv2.Canny(image = img, threshold1=threshold1, threshold2=threshold2)
plt.imshow(edge)
```

```python
threshold1 = 150
threshold2 = 200


edge = cv2.Canny(image = img, threshold1=threshold1, threshold2=threshold2)
plt.imshow(edge)
```

```python
med_value = np.median(img)
med_value
```

```python
lower = int(max(0,0.7*med_value))
upper = int(min(255,1.3*med_value))

edge = cv2.Canny(image = img, threshold1=lower, threshold2=upper)
plt.imshow(edge)
```

```python
# The above did not help at all so we will do the step1, which is first blurr the image
blurred_image = cv2.blur(img,ksize = (5,5))
edge = cv2.Canny(image = blurred_image, threshold1=lower, threshold2=upper)
plt.imshow(edge)
```

```python
blurred_image = cv2.blur(img,ksize = (5,5))
edge = cv2.Canny(image = blurred_image, threshold1=lower, threshold2=upper+50)
plt.imshow(edge)
```

```python
blurred_image = cv2.blur(img,ksize = (5,5))
edge = cv2.Canny(image = blurred_image, threshold1=lower-10, threshold2=upper+80)
plt.imshow(edge)
```
