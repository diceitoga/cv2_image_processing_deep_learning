# Providing own custom seed with Watershed algo...

```python
# This means you can click on the seed. 
road = cv2.imread('./original/Computer-Vision-with-Python/DATA/road_image.jpg')
#plt.imshow(sep_coins)
road_copy = road.copy()
display(road_copy)
print('shape: ' + str(road.shape))
```

```python
img_size  = road.shape[:2]
marker_image = np.zeros(img_size, dtype = np.int32)
segments = np.zeros(road.size, dtype = np.uint8)
```

# Now we will use color mapping

```python
from matplotlib import cm   # cm =  color map
```

```python
cm.tab10(0)
```

```python
# (0.12156862745098039 (red), 0.4666666666666667 (green), 0.7058823529411765(blue), 1.0(alpha))
def create_rgb(i):
    return tuple(np.array(cm.tab10(i)[:3])*255)
```

```python
colors = []

for i in range(10):
    colors.append(create_rgb(i))
    
colors   #each line or truple maps to a color and there are 10 of them...
```

```python
# call back

# Global variables
# Color choice
n_markers = 10
current_marker = 1 
# checks if marks is updated by watershed Algo
marks_updated = False


# Call back functions


# while loop (While True)
```

```python
def mouse_callback(event, x,y, flags, param):
    global marks_updates
    # markers passed to the watershed later on
    if event == cv2.EVENT_FLAG_LBUTTON():
        cv2.circle(marker_image, (x,y),10,(current_marker),-1)  #-1 is solid
    #what the user will see on the 'road' image
        cv2.circle(road_copy, (x,y),10,colors[current_marker],-1)
        
        marks_updates = True
           
```

```python
# while loop listener

cv2.namedWindow('Road Image')
cv2.setMouseCallback('Road Image', mouse_callback)

while True:
    cv2.imshow('Watershed Segment', segments)
    cv2.imshow('Road Image', road_copy)
    
    #close all windows when 'escape key' or 27
    k == cv2.waitKey(1)
        #kill 
    if k ==27:
        break
    
    # clear all of the colors (resets to default) if and when the user presses 'c'
    elif k == order('c'):
        road_copy = road.copy()
        marker_image = np.zeros(road.shape[:2],dtype = np.int32)
        segments = np.zeros(road.shape, dtype= np.uint8)
        
    elif k>0 and char(k).isdigit():
        current_marker = int(char(k))
        
        
    #update color choice
    if marks_updated:
        marker_image_copy = marker_image.copy()
        cv2.watershed(road, marker_image_copy)
        segments = np.zeros(road.shape,dtype = np.uint8)
        
        for color_ind in range(n_markers):
            # coloring segments, numpy call
            segments[marker_image_copy==(color_index)] = colors[color_ind]
            
            
            
        
        
    
    
cv2.destroyAllWindows()
```
