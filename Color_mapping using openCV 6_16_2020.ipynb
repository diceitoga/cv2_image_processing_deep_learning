# Color_mapping using openCV 6_16_2020

```python
'''
HSL model or Hugh, Saturation, lightness

HSV model has Hugh, Saturation, and Value

cvtColor funciton()

'''
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./original/Computer-Vision-with-Python/DATA/00-puppy.jpg')
plt.imshow(img)
```

```python
new_img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
# Now we do this for HSV 

new_HSV= cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(new_HSV)
```

```python
# now for HSL
new_HSL= cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(new_HSL)
```
